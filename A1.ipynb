## Observations on  the experiemnt

* The bias of the constant estimator is higher than the bias of the linear estimator. 
    This means that the difference between the truth, regression function, and the constant estimator is bigger than the difference
    between the truth and linear estimator.
    
* The opposite happens in variance. the variance is higher between the truth and the linear estimator than 
    it is between the truth and the constant estimator.

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
from itertools import product 
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd
#itertools.product
  
def rSubset(arr, r): 
    return list(product(arr, repeat=r))

options = [1, -1]
data_sets = rSubset(options, 5)
print(len(data_sets))
print("random values for coin flips")
print(data_sets)
print(" ")


x_values = [1,2,3,4,5]
y_values = []
final_data_set =[]
for data_set in data_sets:
    for i in range(5):
        y =  x_values[i] + data_set[i]
        y_values.append(y)
    
    final_data_set.append(y_values) 
    y_values = [] 
print("final Y values data set")
print(final_data_set)
    
```

```python
y_results_constants = []
expectation_f_hat = 0
expectation_f_hat_squared = 0
for data_set in final_data_set:
    addition = 0 
    for y in data_set:
        addition +=y
    c = addition / 5
    c_squred = c * c
    expectation_f_hat +=c
    expectation_f_hat_squared += c_squred
    y_results_constants.append(c)
print(" ")
print("The f hat ,constant estimator, values for each training set")
print(y_results_constants)
print("")
expectation_f_hat = expectation_f_hat/32
print("expectation_f_hat of any x =", expectation_f_hat)
expectation_f_hat_squared = expectation_f_hat_squared /32
print("expectation_f_hat squared of any x =", expectation_f_hat_squared)

```

```python

x = [1,2,3,4,5]
y = []
i=0
axs = axs.ravel()
for y_values in final_data_set:
    #y = [y_value] * 5
    plt.figure(i)
    plt.scatter(x,y_values)
    
    

    regr = linear_model.LinearRegression()  
    x = np.array(x).reshape(-1,1)
    y_values = np.array(y_values)
    regr.fit(x, y_values)
     # Make predictions
    y_pred = regr.predict(x)
    print("predected y", y_pred)

    # Plot outputs
    plt.scatter(x, y_values,  color='black')
    plt.plot(x, y_pred, color='blue', linewidth=3)
    constant_estimator = [y_results_constants[i]]*5
    plt.plot(x, constant_estimator, color='red', linewidth=2)
    plt.show()
    
    print(" ")
    # computer bias(f_hat(x1)) = E[f_hat(x1)] - f(x1)
    bias = expectation_f_hat - y_values[0]
    print("Bias of f_hat(x1) = ",expectation_f_hat, " - ", y_values[0], " = " ,bias )

    
    # computer variance(f_hat(x1)) = E[f_hat(x1)^2] - E[f(x1)]^2
    variance =  expectation_f_hat_squared - (expectation_f_hat * expectation_f_hat)
    print("Variance of f_hat(x1) = ",expectation_f_hat_squared, " - ", (expectation_f_hat * expectation_f_hat), " = " ,variance )
    
    
    # computer Bias(f_hat(x1)) = E[f_hat(x1)^2] - E[f(x1)]^2 where f_hat is the linear estimator,f is the regression func
    bias = y_pred[0] - y_values[0]
    print("Bias of linear_estimator(x1) = ",y_pred[0], " - ", y_values[0], " = " ,bias )
    
    # computer variance f_hat(x1) = E[f_hat(x1)^2] - E[f(x1)]^2 where f_hat is the linear estimator,f is the regression func
    variance = (y_pred[0] * y_pred[0]) - (y_values[0] * y_values[0])
    print("Variance of linear_estimator(x1) = ",(y_pred[0] * y_pred[0]), " - ", (y_values[0] * y_values[0]), " = " ,variance )
    
    
    print(" ")
    i+=1
    print(" ----------------------------------------------------------------------------------------------------------")

    
```
