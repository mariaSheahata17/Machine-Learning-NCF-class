```python
import math
from random import random
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
# 1 - heads
# 0 - tails
N_FLIPS = 20
x_axis = [num for num in range(N_FLIPS)]
P0 = 0.5
heads =[]
tails= []
repeat = 20
final_heads_counts =[]
final_tails_counts =[] 
for i in range(repeat):
    observations = [int(random()<P0) for _ in range(N_FLIPS)]    
    for output in observations:
        if output == 1:
            heads.append(output)
        else:
            tails.append(output)   
    final_heads_counts.append(len(heads))
    final_tails_counts.append(len(tails))
    heads = []
    tails = []
#print(final_heads_counts)
#print(final_tails_counts)
```

```python
#plot bar plot

# heads counts
plt.bar(x_axis, final_heads_counts, align='center')
# naming the x-axis 
plt.xlabel('nth experiment') 
# naming the y-axis 
plt.ylabel('Total num of Heads') 
# plot title 
plt.title('20 repitetion of 20 coin flips') 
plt.show()

# tails counts
plt.bar(x_axis, final_tails_counts, align='center')
# naming the x-axis 
plt.xlabel('nth experiment') 
# naming the y-axis 
plt.ylabel('Total num of Tails') 
# plot title 
plt.title('20 repitetion of 20 coin flips') 
plt.show()
```
