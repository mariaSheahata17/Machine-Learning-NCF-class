Y = total amount of change in $$, from
X 1 = number of coins and
X 2 = number of pennies, nickels, and dimes. Report the estimated parameters of the model.

100 pennies, 20 nickels, 10 dimes, or 4 quarters; each = 1 dollar.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


df = pd.read_csv('coins.csv')
df.head()
```

```python
# adding two columns to the table
# one for the interaction term since these variables are correlated
# one for the sum of coins

df['interaction'] = df['Pennies'] * df['Nickels']* df['Dimes']
df['total'] = df['Pennies'] + df['Nickels'] + df['Dimes']
df['change'] = df['Pennies']/100 + df['Nickels']/20 + df['Dimes']/10
df.head()

```

```python

# define the multiple Linear regression model

y = df['change']
x = df[['total','interaction']]


linear_regress = LinearRegression()
linear_regress.fit(x,y)
print(linear_regress.coef_)
y_pred = linear_regress.predict(x)
```
